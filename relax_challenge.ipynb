```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
users = pd.read_csv('./takehome_users.csv', encoding='latin-1')
engagement = pd.read_csv('./takehome_user_engagement.csv')
```

```python
users.head()
```

```python
engagement.head()
```

```python
engagement.time_stamp = engagement.time_stamp.astype(str).str[:10]
```

```python
engagement.head()
```

```python
engagement.time_stamp = pd.to_datetime(engagement.time_stamp)
```

```python
engagement = engagement.groupby(['user_id','time_stamp']).agg({'visited':np.sum})
```

```python
engagement.head()
```

```python
df=engagement.reset_index().groupby('user_id').resample('W',on='time_stamp').agg('sum')
```

```python
df.head()
```

```python
df.rename(columns={'user_id':'id'},inplace=True)
df = df.reset_index().drop('id',axis=1)
```

```python
df = df[df.visited>=3]
```

```python
df.head()
```

```python
users['active'] = np.where(users['object_id'].isin(df.user_id) , 1, 0)
```
